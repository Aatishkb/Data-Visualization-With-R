# **Data Visualization with R Programming**
- Data visualization is a technique used to deliver insights from data using visual cues such as graphs, charts, maps, and more. It helps in intuitive and easy understanding of large quantities of data, thereby aiding in better decision-making

- Types of Data Visualizations in R

- 1. Scatter Plot
- 2. Line Plot
- 3. Bar Plot
- 4. Histogram
- 5. Box Plot
- 6. Violin Plot
- 7. Heatmap
- 8. Density Plot
- 9. Pie Chart
- 10. Faceted Plot

## **1. Scatter Plot**
- A scatter plot displays the relationship between two continuous variables.

```r
# Load the ggplot2 package
library(ggplot2)
# Sample data
data <- data.frame(x = rnorm(100), y = rnorm(100))

# Scatter plot
ggplot(data, aes(x = x, y = y)) +
  geom_point() +
  labs(title = "Scatter Plot", x = "X-Axis", y = "Y-Axis")
```

## **2. Line Plot**
- A line plot is used to visualize trends over time or ordered categories.

```r
# Sample data
data <- data.frame(x = 1:100, y = cumsum(rnorm(100)))

# Line plot
ggplot(data, aes(x = x, y = y)) +
  geom_line() +
  labs(title = "Line Plot", x = "X-Axis", y = "Y-Axis")
```

## **3. Bar Plot**
- Bar plots are useful for displaying the distribution of categorical data.

```r
# Sample data
data <- data.frame(category = c("A", "B", "C"), count = c(10, 23, 15))

# Bar plot
ggplot(data, aes(x = category, y = count)) +
  geom_bar(stat = "identity") +
  labs(title = "Bar Plot", x = "Category", y = "Count")
```

## **4. Histogram**
- Histograms depict the distribution of a continuous variable by dividing it into bins.

```r
# Sample data
data <- data.frame(values = rnorm(1000))

# Histogram
ggplot(data, aes(x = values)) +
  geom_histogram(binwidth = 0.5, fill = "blue", color = "black") +
  labs(title = "Histogram", x = "Values", y = "Frequency")
```

## **5. Box Plot**
- Box plots summarize the distribution of a continuous variable and identify potential outliers.

```r
# Sample data
data <- data.frame(category = rep(c("A", "B", "C"), each = 100),
                   values = c(rnorm(100, mean = 5), rnorm(100, mean = 6), rnorm(100, mean = 7)))

# Box plot
ggplot(data, aes(x = category, y = values)) +
  geom_boxplot() +
  labs(title = "Box Plot", x = "Category", y = "Values")
```

##**6. Violin Plot**

- Violin plots combine aspects of box plots and density plots to show the distribution of the data across different categories.

```r
# Sample data
data <- data.frame(category = rep(c("A", "B", "C"), each = 100),
                   values = c(rnorm(100, mean = 5), rnorm(100, mean = 6), rnorm(100, mean = 7)))

# Violin plot
ggplot(data, aes(x = category, y = values)) +
  geom_violin(fill = "lightblue") +
  labs(title = "Violin Plot", x = "Category", y = "Values")
```

## **7. Heatmap**

- Heatmaps represent data values as colors in a matrix layout, useful for visualizing correlations or patterns.

```r
# Sample data
data <- data.frame(x = rep(1:10, each = 10),
                   y = rep(1:10, times = 10),
                   z = rnorm(100))
# Heatmap
ggplot(data, aes(x = x, y = y, fill = z)) +
  geom_tile() +
  labs(title = "Heatmap", x = "X-Axis", y = "Y-Axis")
```

## **8. Density Plot**

- Density plots show the distribution of a continuous variable, similar to histograms but smoothed.

```r
# Sample data
data <- data.frame(values = rnorm(1000))

# Density plot
ggplot(data, aes(x = values)) +
  geom_density(fill = "lightgreen") +
  labs(title = "Density Plot", x = "Values", y = "Density")
```

## **9. Pie Chart**

- Pie charts represent the proportion of categories in a whole.

```r
# Sample data
data <- data.frame(category = c("A", "B", "C"), count = c(10, 23, 15))

# Pie chart
ggplot(data, aes(x = "", y = count, fill = category)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar("y") +
  labs(title = "Pie Chart")
```

## **10. Faceted Plot**

- Faceting allows the creation of multiple plots based on the subsets of a dataset, useful for comparing distributions across categories.

```r
# Sample data
data <- data.frame(category = rep(c("A", "B", "C"), each = 100),
                   x = rep(1:100, 3),
                   y = c(cumsum(rnorm(100)), cumsum(rnorm(100)), cumsum(rnorm(100))))

# Faceted plot
ggplot(data, aes(x = x, y = y)) +
  geom_line() +
  facet_wrap(~ category) +
  labs(title = "Faceted Plot", x = "X-Axis", y = "Y-Axis")
```
